### Create WMI Mapping
Builds a dictionary to link WMI codes to valid 'make' values using existing data where 'make' is known 
This approach helps fill in missing car manufacturer names using the WMI code from the VIN.

```python
# Converts both vin and make to string type to avoid type errors in later operations.

df['vin'] = df['vin'].astype(str)
df['make'] = df['make'].astype(str)

# Extract the WMI (World Manufacturer Identifier)

df['wmi']=df['vin'].str.slice(0,3)


```

```python
# Count Initial 'NO_VALUE' Entries in 'make'

initial_no_value_count = df[df['make'] == 'NO_VALUE'].shape[0]
print("Initial NO_VALUE count in 'make':", initial_no_value_count)
```

### Build WMI-to-Make Mapping

- Filters the DataFrame to exclude rows where vin is "NoValue" or make is "NO_VALUE".
- Selects the columns wmi and make.
- Uses .set_index('wmi')['make'] to convert the DataFrame into a dictionary {wmi: make}.
- This dictionary (wmi_mapping) allows us to look up the make based on the WMI.

```python
wmi_mapping = (
    df.loc[(df['vin'] != 'NoValue') & (df['make'] != 'NO_VALUE'), ['wmi', 'make']]
    .set_index('wmi')['make']
    .to_dict()
)

print("Number of valid mappings in WMI_mapping:", len(wmi_mapping))
```

```python
wmi_mapping
```

```python
print("Number of valid mappings in WMI_mapping:", len(wmi_mapping))
```

```python
def map_make(row):
    if row['make'] == 'NO_VALUE':  # Only update if 'make' is missing
        wmi_value = row['wmi']  
        if wmi_value in wmi_mapping:  # Check if WMI is in mapping
            return wmi_mapping[wmi_value]  # Replace with mapped value
    return row['make']
```

```python
df['make'] = df.apply(map_make, axis=1)
```

```python
updated_no_make_count = df[df['make'] == 'NO_VALUE'].shape[0]
print("NO_VALUE count after update:", updated_no_make_count)
```

### Create VDS Mapping
Builds a dictionary to link VDS codes to valid 'model' and 'trim' values using existing data where 'model' and 'trim' is known. This approach helps fill in missing car model and trim names using the VDS code from the VIN.

```python
df['vin'] = df['vin'].astype(str)
df['model'] = df['model'].astype(str)

df['vds'] = df['vin'].str.slice(3, 8)


initial_no_model_count = df[df['model'] == 'no_model'].shape[0]
print("Initial NO_MODEL count in 'model':", initial_no_model_count)

vds_mapping = (
    df.loc[(df['vin'] != 'NoValue') & (df['model'] != 'no_model'), ['vds', 'model']]
    .set_index('vds')['model']
    .to_dict()
)

print("Number of valid mappings in vds_mapping:", len(vds_mapping))

def map_make(row):
    if row['model'] == 'no_model':
        vds_value = row['vds']  
        if vds_value in vds_mapping:
            return vds_mapping[vds_value]
    return row['model']


df['model'] = df.apply(map_make, axis=1)


updated_no_model_count = df[df['model'] == 'no_model'].shape[0]
print("NO_MODEL count after update:", updated_no_model_count)


remaining_nan_count = df['model'].isna().sum()
print("Remaining NaN values in 'model':", remaining_nan_count)


df.drop(columns=['vds'], inplace=True)
```

### Extract VDS Segment
The 5th character of the VIN is stored in a new column called 'vds'.

### Create VDS-Body Mapping
Builds a dictionary to link VDS codes to valid 'body' values using existing data where 'body' is known This approach helps fill in missing car body types using a character from the VDS code in the VIN.



```python
df['vin'] = df['vin'].astype(str)
df['body'] = df['body'].astype(str)

df['vds'] = df['vin'].str.slice(4, 5)


initial_no_body_count = df[df['body'] == 'no_body'].shape[0]
print("Initial NO_BODY count in 'body':", initial_no_body_count)

vds_body = (
    df.loc[(df['vin'] != 'NoValue') & (df['body'] != 'no_body'), ['vds', 'body']]
    .set_index('vds')['body']
    .to_dict()
)

print("Number of valid mappings in vds_body:", len(vds_body))

def map_make(row):
    if row['body'] == 'no_body':
        vds_value = row['vds']  
        if vds_value in vds_body:
            return vds_body[vds_value]
    return row['body']


df['body'] = df.apply(map_make, axis=1)


updated_no_body_count = df[df['body'] == 'no_body'].shape[0]
print("NO_BODY count after update:", updated_no_body_count)


remaining_nan_count = df['body'].isna().sum()
print("Remaining NaN values in 'body':", remaining_nan_count)


df.drop(columns=['vds'], inplace=True)
```

### Extract VDS Segment
The 7th character of the VIN is stored in a new column called 'vds'.

### Create VDS-Transmission Mapping
Builds a dictionary to link VDS codes to valid 'transmission' values using existing data where 'transmission' is known. This approach helps fill in missing transmission types using a character from the VDS code in the VIN.



```python
df['vin'] = df['vin'].astype(str)
df['transmission'] = df['transmission'].astype(str)

df['vds'] = df['vin'].str.slice(6, 7)


initial_no_transmission_count = df[df['transmission'] == 'no_transmission'].shape[0]
print("Initial no_transmission count in 'body':", initial_no_transmission_count)

vds_transmission = (
    df.loc[(df['vin'] != 'NoValue') & (df['transmission'] != 'no_transmission'), ['vds', 'transmission']]
    .set_index('vds')['transmission']
    .to_dict()
)

print("Number of valid mappings in vds_transmission:", len(vds_transmission))

def map_make(row):
    if row['transmission'] == 'no_transmission':
        vds_value = row['vds']  
        if vds_value in vds_transmission:
            return vds_transmission[vds_value]
    return row['transmission']


df['transmission'] = df.apply(map_make, axis=1)


updated_no_transmission_count = df[df['transmission'] == 'no_transmission'].shape[0]
print("no_transmission count after update:", updated_no_transmission_count)


remaining_nan_count = df['transmission'].isna().sum()
print("Remaining NaN values in 'transmission':", remaining_nan_count)


df.drop(columns=['vds'], inplace=True)
```

```python
df['condition'] = pd.to_numeric(df['condition'], errors='coerce')

def replace(row):
    if row['condition'] == 5.0:
      return 'Excellent'
    elif (row['condition'] <= 4.9) and (row['condition'] >= 4.0):
        return 'Good'
    elif (row['condition'] <= 3.9) and (row['condition'] >= 3.0):
        return 'Fair'
    elif (row['condition'] <= 2.9) and (row['condition'] >= 2.0):
        return 'Poor'
    elif row['condition'] < 2.0:
        return 'Needs major maintenance'
    else :
        return row['condition']


df['condition']=df.apply(replace, axis=1)
        
```

```python
def impute_condition(row):
    if (row['condition']=='no_condition'):
        if row['sellingprice'] > 169000.0:
            return 'Good'
        elif row['sellingprice'] <= 169000.0 and row['sellingprice']>165000.0:
            return 'Fair'
        elif (row['sellingprice'] <= 165000.0) and (row['sellingprice']>=117000.0):
            return 'Poor'
        else:
            return 'Needs major maintenance'
    return row['condition']

df['condition'] = df.apply(impute_condition, axis=1)
```

```python
outliers={}
df['odometer'] = pd.to_numeric(df['odometer'], errors='coerce')
numerical_cols = ['odometer','mmr']
for col in numerical_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    print(f"\nOutliers in {col}:\n", outliers[col])
```
